```python
%matplotlib inline
import pandas as pd
from plotnine import *
```

## Step 1:

```python
df = pd.read_csv("CSV Files/DataRecord_3_DLC_Weight_File_05Feb2019 (1).csv")
df
```

Conclusion:

## Step 2:

```python
adults = df.drop(df.index[df["Age_Category"] != "adult"])
adults
```

```python
ggplot(adults, aes(x = "Sex", y = "Weight_g")) + geom_boxplot()\
+ xlab("Sex") + ylab("Weight (g)") + ggtitle("Weight By sex")
```

Conclusion: All things considered there is very little difference between the weights of the lemurs by sex amongst all of the different species. The extremes of the female is higher but the median and average are very similar.

## Step 3:

```python
ggplot(adults, aes(x = "Taxon", y = "Weight_g")) + geom_boxplot()\
+ xlab("Species") + ylab("Weight (g)") + ggtitle("Weight By Species")\
+ theme(axis_text_x=element_text(rotation=90))
```

Conclusion: Not only is the median and average of the sifaka lemurs the highest out of all other species, it also by far has the highest extreme weight values amongst all of the other species. Sifaka lemurs consistently outweigh every other species.

## Step 4:

```python
sifaka = adults.drop(adults.index[adults["Taxon"] != "PCOQ"])
sifaka
```

```python
ggplot(sifaka, aes(x="AgeAtWt_y", y="Weight_g", color="Age_Category")) \
+ geom_point(alpha=0.1)\
+ geom_smooth(se=False, method="lm")\
+ xlab("Weight at specific age") + ylab("Weight (g)") + ggtitle("Weight By Age among adult Sifaka lemurs")
```

Conclusion: Generally the graph trends downwards as age increases among the sifaka lemurs meaning they usually weigh the most at early points in their life and lose that mass as their life continues. There are some outliers very early on in age sometimes having more than double the average weight for an adult sifaka lemur.

Conclusion:

## Step 7:

```python
reduced = df2[['Latin_Abbreviation', 'R_Ratio_MtoF_DLCBirths', 'R_Mean_LitterSize', 'R_Expected_Gestation_d',
              'R_Peak_Birth_Month', 'M_Mean_All_AdultWeight_g', 'M_Mean_All_NeonateWeight_g', 'M_Mean_All_YngAdultWeight_g',
              'L_Median_All_Longevity_gt30d_y', 'L_Pct_All_InfMort_lt30d']]
reduced
```

```python
reduced = reduced.dropna()
reduced
```

## Step 8:

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters = 2, init="random",
           n_init = 10, max_iter=300,
           tol=1e-04, random_state=0)
```

```python
x = reduced.drop(['Latin_Abbreviation'], axis=1)
x = x.values
```

```python
result = km.fit(x)
result.cluster_centers_
```

```python
result.labels_
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(x)
```

```python
pcaReduced = pd.DataFrame(data= principal_components, columns=["PCA 1", "PCA 2"])
pcaReduced
```

```python
pcaReduced["cluster"] = pd.Series(result.labels_).astype("str")
pcaReduced
```

```python
pcaReduced["Latin Abbreviation"] = reduced["Latin_Abbreviation"]
pcaReduced
```

## Step 10:

```python
ggplot(pcaReduced, aes(x="PCA 1", y="PCA 2", color="cluster"))\
+ geom_point()\
+ geom_text(aes(label="Latin Abbreviation"))\
+ xlab("PCA 1") + ylab("PCA 2") + ggtitle("Cluster Plot with named points")
```

Conclusion: This cluster chart showing the different lemur species closely coincides with what was shown in the graph from step 3. The species that were similar to eachother in weight as shown in the box plot from step 3 are plotted nearer eachother in this cluster plot. Given this data the closest to the Sifaka lemurs is probably the Northern giant mouse lemur however noone of the species within the Sifaka's cluster are very close to it all keeping a similar distance away from the species. This is likely to due with the Sifaka's general uniqueness in it's data with no other lemur particularly being similar to the Sifaka especially in something like weight.
